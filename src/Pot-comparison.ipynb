```python
import HeBz 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vectorize the library functions
Vconsvec = np.vectorize(HeBz.Vcons)
VLeevec = np.vectorize(HeBz.Lee2003)
VShirkovvec = np.vectorize(HeBz.Shirkov2024)
LJvec = np.vectorize(HeBz.LennardJones)
```

```python
xplot = np.linspace(-10,10,100)
yplot = np.linspace(-10,10,100)
zplot = np.linspace(-10,10,100)
```

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(2*8,2*5), constrained_layout=True,sharex=True)
line_width = 1.25
plt.gcf().set_dpi(500)
font_size = 13
ax1.plot(xplot,Vconsvec(0,0,zplot), label="Lookup", linewidth=line_width)
ax1.plot(xplot,VLeevec(0,0,zplot), label="Lee 2003", linewidth=line_width)
ax1.plot(xplot,VShirkovvec(0,0,zplot), label="Shirkov 2024", linewidth=line_width)
ax1.plot(xplot,LJvec(0,0,zplot), label="Lennard-Jones", linewidth=line_width)
ax2.plot(xplot,Vconsvec(xplot,0,2.5), label="Lookup", linewidth=line_width)
ax2.plot(xplot,VLeevec(xplot,0,2.5), label="Lee 2003", linewidth=line_width)
ax2.plot(xplot,VShirkovvec(xplot,0,2.5), label="Shirkov 2024", linewidth=line_width)
ax2.plot(xplot,LJvec(xplot,0,2.5), label="Lennard-Jones", linewidth=line_width)
ax2.legend(loc="upper right",frameon=False,fontsize=font_size)
ax1.set_ylim(-200,400)
ax3.plot(xplot,Vconsvec(0,yplot,3), label="Lookup", linewidth=line_width)
ax3.plot(xplot,VLeevec(0,yplot,3), label="Lee 2003", linewidth=line_width)
ax3.plot(xplot,VShirkovvec(0,yplot,3), label="Shirkov 2024", linewidth=line_width)
ax3.plot(xplot,LJvec(0,yplot,3), label="Lennard-Jones", linewidth=line_width)
ax4.plot(xplot,Vconsvec(2,0.6,zplot), label="Lookup", linewidth=line_width)
ax4.plot(xplot,VLeevec(2,0.6,zplot), label="Lee 2003", linewidth=line_width)
ax4.plot(xplot,VShirkovvec(2,0.6,zplot), label="Shirkov 2024", linewidth=line_width)
ax4.plot(xplot,LJvec(2,0.6,zplot), label="Lennard-Jones", linewidth=line_width)
ax4.set_ylim(-200,400)
#plt.rc("font", size=10)  # controls default text sizes
#plt.rc("xtick", labelsize=font_size)  # fontsize of the tick labels
#plt.rc("ytick", labelsize=font_size)  # fontsize of the tick labels
ax3.set_xlabel(r"Distance ($\AA$)", fontsize=font_size)
ax4.set_xlabel(r"Distance ($\AA$)", fontsize=font_size)
ax1.set_ylabel(r"V(0,0,z)(K)", fontsize=font_size)
ax2.set_ylabel(r"V(x,0,2.5)(K)", fontsize=font_size)
ax3.set_ylabel(r"V(0,y,3)(K)", fontsize=font_size)
ax4.set_ylabel(r"V(2,0.6,z)(K)", fontsize=font_size)
# plt.title("CCSD(T) Potential Energy Surface")
#plt.legend(loc="lower right", fontsize=font_size)
plt.savefig("../figures/Comparison.pdf")
```
