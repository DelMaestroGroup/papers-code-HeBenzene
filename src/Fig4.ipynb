```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.image as mpimg
from matplotlib.patches import ConnectionPatch
```

```python
datafixed = np.loadtxt('../data/pimc-fixed/estimator-u-reduce.dat')
dataLee = np.loadtxt('../data/pimc-Lee/estimator-u-reduce.dat')
datagp = np.loadtxt('../data/pimc-gp/estimator-u-reduce.dat')
dataShirkov = np.loadtxt('../data/pimc-Shirkov/estimator-u-reduce.dat')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5),constrained_layout=True)
plt.gcf().set_dpi(500)
ax.errorbar(datafixed[:,0],datafixed[:,19],yerr=datafixed[:,20],marker='.',linestyle='--',label=r'Lennard-Jones')
ax.errorbar(dataLee[:,0],dataLee[:,19],yerr=dataLee[:,20],marker='.',linestyle=':',label=r'Lee 2003')
ax.errorbar(dataShirkov[:,0],dataShirkov[:,19],yerr=dataShirkov[:,20],linestyle='-.',marker='.',label=r'Shirkov 2024')
ax.errorbar(datagp[:,0],datagp[:,19],yerr=datagp[:,20],marker='.',linestyle='-',label=r'This work')

p1 = [-45,2.0]
p2 = [-29,4.6]
p3 = [-12,22.8]
ax_inset1 = ax.inset_axes([0.03, 0.15, 0.3, 0.3]) 
img1 = mpimg.imread("../figures/benzene_densitymu45.png")  
ax_inset1.imshow(img1)
ax_inset1.set_xticks([])
ax_inset1.set_yticks([])
ax_inset1.set_xlim(400,1500)
ax_inset1.set_ylim(1100,100)
#ax_inset.axis('off') 
# Get corners of the inset in data coordinates
corners = [
    (ax_inset1.get_xlim()[1], ax_inset1.get_ylim()[1]), 
    (ax_inset1.get_xlim()[1], ax_inset1.get_ylim()[0])   
]

# Draw lines from inset corners to the highlighted point
for corner in corners:
    con = ConnectionPatch(
        xyA=(p1[0], p1[1]), coordsA=ax.transData,
        xyB=corner, coordsB=ax_inset1.transData,
        arrowstyle="-", color="gray", lw=0.5
    )
    ax_inset1.add_artist(con)
ax_inset2 = ax.inset_axes([0.35, 0.35, 0.3, 0.3]) 
img2 = mpimg.imread("../figures/benzene_densitymu29.png")  
ax_inset2.imshow(img2)
ax_inset2.set_xticks([])
ax_inset2.set_yticks([])
ax_inset2.set_xlim(400,1500)
ax_inset2.set_ylim(1100,100)
#ax_inset.axis('off') 
# Get corners of the inset in data coordinates
corners = [
    (ax_inset2.get_xlim()[1], ax_inset2.get_ylim()[1]), 
    (ax_inset2.get_xlim()[0], ax_inset2.get_ylim()[0])   
]

# Draw lines from inset corners to the highlighted point
for corner in corners:
    con = ConnectionPatch(
        xyA=(p2[0], p2[1]), coordsA=ax.transData,
        xyB=corner, coordsB=ax_inset2.transData,
        arrowstyle="-", color="gray", lw=0.5
    )
    ax_inset2.add_artist(con)
ax_inset3 = ax.inset_axes([0.58, 0.67, 0.3, 0.3]) 
img3 = mpimg.imread("../figures/benzene_densitymu12.png") 
ax_inset3.imshow(img3)
ax_inset3.set_xticks([])
ax_inset3.set_yticks([])
ax_inset3.set_xlim(400,1500)
ax_inset3.set_ylim(1100,100)
#ax_inset.axis('off') 
# Get corners of the inset in data coordinates
corners = [
    (ax_inset3.get_xlim()[1], ax_inset3.get_ylim()[1]), 
    (ax_inset3.get_xlim()[1], ax_inset3.get_ylim()[0])   
]

# Draw lines from inset corners to the highlighted point
for corner in corners:
    con = ConnectionPatch(
        xyA=(p3[0], p3[1]), coordsA=ax.transData,
        xyB=corner, coordsB=ax_inset3.transData,
        arrowstyle="-", color="gray", lw=0.5
    )
    ax_inset3.add_artist(con)

ax.set_xlabel(r'$\mu$ (K)')
ax.set_ylabel(r'$\langle N\rangle$')
ax.set_xlim(-85,-12)
ax.set_ylim(-0.5,27)
plt.legend()
plt.savefig('../figures/Fig4.pdf')
```
