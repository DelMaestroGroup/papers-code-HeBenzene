```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
datafixed = np.loadtxt('../data/pimc-fixed/estimator-u-reduce.dat')
dataLee = np.loadtxt('../data/pimc-Lee/estimator-u-reduce.dat')
datagp = np.loadtxt('../data/pimc-gp/estimator-u-reduce.dat')
dataShirkov = np.loadtxt('../data/pimc-Shirkov/estimator-u-reduce.dat')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
plt.gcf().set_dpi(500)
ax.errorbar(datafixed[:,0],datafixed[:,19],yerr=datafixed[:,20],label=r'Lennard-Jones')
ax.errorbar(dataLee[:,0],dataLee[:,19],yerr=dataLee[:,20],label=r'Lee 2003')
ax.errorbar(dataShirkov[:,0],dataShirkov[:,19],yerr=dataShirkov[:,20],label=r'Shirkov 2024')
ax.errorbar(datagp[:,0],datagp[:,19],yerr=datagp[:,20],label=r'This work')

ax.set_xlabel(r'$\mu$ (K)')
ax.set_ylabel('N')
ax.set_xlim(-85,-12)
plt.legend(frameon=False)
plt.savefig('../figures/Fig4.svg')
```
