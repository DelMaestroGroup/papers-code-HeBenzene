```python
import numpy as np
import matplotlib.pyplot as plt
import HeBz
```

```python
Shirkvec = np.vectorize(HeBz.Shirkov2024)
VVec = np.vectorize(HeBz.V)
```

```python
#Find the minimum
zp = np.linspace(2,7,500)
minind = np.argmin(Shirkvec(0,0,zp))
minp = zp[minind]
print(minp)
```

```python
xp = np.linspace(-5,5,100)
yp = np.linspace(-5,5,100)
X, Y = np.meshgrid(xp, yp)
Z = VVec(X,Y,minp)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
plt.gcf().set_dpi(500)
im = ax.imshow(Z,interpolation='none',extent=[-5,5,-5,5])
font_size = 13
plt.rc("font", size=10) 
ax.set_xlabel(r'$x$ ()', size=font_size)
ax.set_ylabel(r'$y$ ()', size=font_size)
cbar = fig.colorbar(im, ax=ax)
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel(r'$V(x,y,3.1423)$ (K)', rotation=270)
#ax.legend()

plt.savefig('../figures/Poturf.svg')
```
