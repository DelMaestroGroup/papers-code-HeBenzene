```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.style.use('aps')
mpl.rcParams["figure.figsize"] = [3.4039, 2.10373]
```

```python
f2 = open('estimator-u-reduce.dat','r')
lines = f2.readlines()
u = np.array([])
N = np.array([])
delN = np.array([])
for line in lines[1:]:
    p = line.split()
    u = np.append(u,float(p[0]))
    N = np.append(N,float(p[19]))
    delN = np.append(delN,float(p[20]))
f2.close()
ind = np.argsort(u)
u = u[ind]
N = N[ind]
print(u)
```

```python
figsize = plt.rcParams['figure.figsize']
fig,ax = plt.subplots(figsize=(figsize[0],figsize[1]), constrained_layout=True)
ax.plot(u,N,label=r'T = 2.0 K')
ax.set_xlabel(r'$\mu$ (K)')
ax.set_ylabel('N')
plt.legend(frameon=False)
plt.savefig('BenzeneNmu.pdf')
```
