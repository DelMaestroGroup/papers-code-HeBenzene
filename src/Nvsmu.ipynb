```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
f2 = open('../data/fixed/estimator-u-reduce.dat','r')
lines = f2.readlines()
u = np.array([])
N = np.array([])
delN = np.array([])
for line in lines[1:]:
    p = line.split()
    u = np.append(u,float(p[0]))
    N = np.append(N,float(p[19]))
    delN = np.append(delN,float(p[20]))
f2.close()
ind = np.argsort(u)
ufixed = u[ind]
Nfixed = N[ind]
print(u)
f2.close()
f2 = open('../data/fit/estimator-u-reduce.dat','r')
lines = f2.readlines()
u = np.array([])
N = np.array([])
delN = np.array([])
for line in lines[1:]:
    p = line.split()
    u = np.append(u,float(p[0]))
    N = np.append(N,float(p[19]))
    delN = np.append(delN,float(p[20]))
f2.close()
ind = np.argsort(u)
uVfit = u[ind]
NVfit = N[ind]
print(u)
f2.close()
f2 = open('../data/lookup/estimator-u-reduce.dat','r')
lines = f2.readlines()
u = np.array([])
N = np.array([])
delN = np.array([])
for line in lines[1:]:
    p = line.split()
    u = np.append(u,float(p[0]))
    N = np.append(N,float(p[19]))
    delN = np.append(delN,float(p[20]))
f2.close()
ind = np.argsort(u)
ulookup = u[ind]
Nlookup = N[ind]
print(u)
```

```python
figsize = plt.rcParams['figure.figsize']
fig,ax = plt.subplots(figsize=(8,5), constrained_layout=True)
line_width = 1.25
plt.gcf().set_dpi(500)
font_size = 13
ax.plot(ufixed,Nfixed,label=r'Lennard-Jones',linewidth=line_width)
ax.plot(uVfit,NVfit,label=r'Lee 2003',linewidth=line_width)
ax.plot(ulookup,Nlookup,label=r'lookup',linewidth=line_width)
ax.set_xlabel(r'$\mu$ (K)',fontsize=font_size)
ax.set_ylabel('N',fontsize=font_size)
ax.set_xlim(-85,-12)
plt.legend(handlelength=1,frameon=False,fontsize=font_size)
plt.savefig('../figures/Nmu-comp.pdf')
```
